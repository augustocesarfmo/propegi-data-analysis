```python
#Visualizacao crua dos Dados


import pandas as pd
import json

caminho_arquivo = "input/Financas.json"

# Abri o JSON
with open(caminho_arquivo, encoding="utf-8") as f:
    dados = json.load(f)

# Converter para DataFrame
df = pd.DataFrame(dados)

# Visualizar os primeiros dados
df.head()
```

```python
# Exportar os dados tratados
df.to_csv('output/dados_tratados.csv', index=False, encoding='utf-8-sig')
```
